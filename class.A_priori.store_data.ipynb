L'algorithme Apriori utilise des jeux d'éléments (items) fréquents pour générer des règles d'association (rule), et il est conçu pour fonctionner sur les bases de données qui contiennent des transactions.

algo apriori determine a quel point deux objets sont conectés fortement ou faiblement.
Il est principalement utilisé pour l'analyse du panier de consommation et aide à trouver 
les produits qui peuvent etre acheté ensemble. Il peut également etre utilisé 
dans le domaine de la santé pour trouver des reactions medicamenteuses pour les patients.

Qu'est-ce que l'ensemble d'éléments fréquents (frequent itemset) ?

L'ens d'éléments fréquents sont les éléments dont le support est supérieure à la valeur seuil ou au support minimale spécifiée par l'utilisateur. 

Étapes de l'algorithme Apriori
Voici les étapes de l'algorithme apriori :

Étape 1 : Déterminer le support des ens de données (support of Items) dans la base de données transactionnelle et sélectionner la prise en charge (min_support) et la confiance minimales 
(min_confidence).

Étape 2 : Prendre tous les supports de la transaction avec une valeur de support supérieure à la valeur de support minimale ou sélectionnée.

Étape 3 : Trouver toutes les règles (rules) de ces sous-ensembles qui ont une valeur de confiance plus élevée que le seuil ou la confiance minimale.

Étape 4 : Trier les règles par ordre décroissant d'ascenseur (lift).

Fonctionnement de l'algorithme apriori
Nous allons comprendre l'algorithme a priori à l'aide d'un exemple et d'un calcul mathématique :

Exemple : supposons que nous ayons l'ensemble de données suivant qui comporte diverses transactions, et à partir de cet ensemble de données, nous devons trouver les ensembles d'éléments fréquents et générer les règles d'association à l'aide de l'algorithme Apriori :

TID:       T1       T2    T3    T4,      T5      T6     T7     T8         T9
ITEMSETS:  A,B     B,D    B,C   A,B,D    A,C    B,C    A,C     A,B,C,E   A,B,C

sachant: Min_support=2, min_confidence= 50%

Solution:
Étape 1 : Calculer C1 et L1 :
Dans la première étape, nous allons créer un tableau qui contient le nombre de supports (la fréquence de chaque ensemble d'éléments individuellement dans l'ensemble de données) de chaque ensemble d'éléments dans l'ensemble de données donné. Cette tableau est appelée ensemble candidat ou C1.

itemset                A         B        C       D     E
support_count          6          7       5       2     1


Maintenant, nous allons retirer tous les ensembles d'éléments qui ont le plus grand nombre de supports que le support minimum (2). Cela nous donnera le tableau pour l' item fréquent L1.
Étant donné que tous les ensembles d'éléments ont un nombre de supports supérieur ou égal au support minimum, à l'exception du E, l'ensemble d'éléments E sera supprimé.

itemset                A         B        C       D    
support_count          6          7       5       2  


Étape 2 : Génération de candidats C2 et L2 :
Dans cette étape, nous allons générer C2 à l'aide de L1. En C2, nous allons créer le couple des itemsets de L1 sous forme de sous-ensembles.
Après avoir créé les sous-ensembles, nous retrouverons le nombre de supports dans la table de transaction principale des ensembles de données, c'est-à-dire combien de fois ces paires se sont produites ensemble dans l'ensemble de données donné. Ainsi, nous obtiendrons le tableau ci-dessous pour C2 :

itemset                (A,B)     (A,C)    (A,D)      (B,C)    (B,D)   (C,D)    
support_count           4          4       1           4        2       0


Encore une fois, nous devons comparer le nombre de supports C2 avec le nombre de supports minimum, et après la comparaison, l'ensemble d'éléments avec le moins de nombre de supports sera éliminé du tableau C2. Cela nous donnera le tableau ci-dessous pour L2

itemset                (A,B)     (A,C)    (A,D)      (B,C)    (B,D)    
support_count           4          4       1           4        2  

        A, B, C, D

Étape 3 : génération de candidats C3 et L3 :
Pour C3, nous allons répéter les deux mêmes processus, mais maintenant nous allons former la table C3 avec des sous-ensembles de trois ensembles d'éléments ensemble, et calculer le nombre de supports à partir de l'ensemble de données. Cela donnera le tableau ci-dessous :

itemset                (A,B,C)     (B,C,D)   (A,C,D)    (A,B,D)       
support_count           2              0       0           1        

Nous allons maintenant créer la table L3. Comme nous pouvons le voir dans le tableau C3 ci-dessus, il n'y a qu'une seule combinaison d'éléments dont le nombre de supports est égal au nombre de supports minimum. Ainsi, le L3 n'aura qu'une seule combinaison, c'est-à-dire {A, B, C}.

Étape 4 : Recherche des règles d'association pour les sous-ensembles :
Pour générer les règles d'association, nous allons d'abord créer une nouvelle table avec les règles possibles à partir de la combinaison {A, BC}. Pour toutes les règles, nous calculerons la Confiance en utilisant la formule sup( A ^B)/A. Après avoir calculé la valeur de confiance pour toutes les règles, nous exclurons les règles qui ont moins de confiance que le seuil minimum (50%).

Considérez le tableau ci-dessous :

Règles	       Support	Confiance
A ^B → C	2	Sup{(A ^B) ^C}/sup(A ^B)= 2/4=0,5=50%
B^C → A	        2	Sup{(B^C) ^A}/sup(B ^C)= 2/4=0,5=50%
A^C → B	        2	Sup{(A ^C) ^B}/sup(A ^C)= 2/4=0,5=50%
C→ A ^B	        2	Sup{(C^( A ^B)}/sup(C)= 2/5=0.4=40%
A→ B^C	        2	Sup{(A^( B ^C)}/sup(A)= 2/6=0.33=33.33%
B→ B^C    	2	Sup{(B^( B ^C)}/sup(B)= 2/7=0.28=28%

Comme le seuil ou la confiance minimale donné est de 50 %, les trois premières règles A ^B → C, B^C → A et A^C → B peuvent être considérées comme les règles d'association fortes pour le problème donné.

Avantages de l'algorithme Apriori
C'est un algorithme facile à comprendre
Les étapes de jointure et d'élagage de l'algorithme peuvent être facilement implémentées sur de grands ensembles de données.

Inconvénients de l'algorithme Apriori
L'algorithme apriori fonctionne lentement par rapport aux autres algorithmes.
Les performances globales peuvent être réduites car il analyse la base de données plusieurs fois.
La complexité temporelle et spatiale de l'algorithme a priori est de O(2^D ), ce qui est très élevé. Ici D représente la largeur horizontale présente dans la base de données.

2. Formation du modèle Apriori sur l'ensemble de données
Pour entraîner le modèle, nous utiliserons la fonction apriori qui sera importée du package apyroi . Cette fonction renverra les règles pour entraîner le modèle sur l'ensemble de données. Considérez le code ci-dessous 


la fonction apriori renvoie la sortie sous forme de règles. Il prend les paramètres suivants :

transactions : Une liste de transactions.
min_support = Pour définir la valeur flottante minimale du support. Ici, nous avons utilisé 0,003 qui est calculé en prenant 3 transactions par client chaque semaine sur le nombre total de transactions.
min_confidence : Pour définir la valeur de confiance minimale. Ici, nous avons pris 0,2. Il peut être modifié selon le problème de l'entreprise.
min_lift = Pour définir la valeur d'ascenseur minimum.
min_length = Il faut le nombre minimum de produits pour l'association.
max_length = Il prend le nombre maximum de produits pour l'association.


```python
transactions = []
for i in range (0, 7501):
    transactions.append([str(store_data.values[i,j]) for j in range(0, 20)])
association_rules = apriori(transactions, min_support=0.0045, min_confidence=0.2, min_lift=3,min_length=2)
association_results= list(association_rules)
```

Comme nous pouvons le voir, la sortie ci-dessus est sous une forme qui n'est pas facilement compréhensible. Ainsi, nous imprimerons toutes les règles dans un format approprié.

Visualiser la règle, le support, la confiance, la levée (lift) de manière plus claire :

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule:"  + items[0] + "->" + items[1]) 
    print("Support:"  + str(item[1]))  # popularité individuelle (ItemA/TotalTransactions)
    print("Confidence:"  + str(item[2][0][2]))  # Item A induira plus de profit pour Item B (  Item(A+B)/Item(A)  )
    print("Lift:"  + str(item[2][0][3])) # combien Item A induit un profit dans Item B (cofidence/Support)
    print("==================================") 
```
